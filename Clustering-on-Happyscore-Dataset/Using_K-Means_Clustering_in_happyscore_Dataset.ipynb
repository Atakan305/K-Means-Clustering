```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('happyscore_income')

happiness = data['happyScore']
income = data['avg_income']

happiness.max()
```

```python
plt.xlabel('income')
plt.ylabel('happiness')
plt.scatter(income, happiness)
```

```python
data.sort_values('avg_income', inplace = True)

richest = data[data['avg_income'] > 20000]

richest_mean = np.mean(richest['avg_income'])
all_mean = np.mean(data['avg_income'])

print(all_mean, richest_mean)
```

```python
plt.scatter(richest['avg_income'], richest['happyScore'])

plt.text(richest.iloc[0]['avg_income'],
         richest.iloc[0]['happyScore'],
         richest.iloc[0]['country'])

plt.text(richest.iloc[-1]['avg_income'],
         richest.iloc[-1]['happyScore'],
         richest.iloc[-1]['country'])
```

```python
plt.scatter(richest['avg_income'], richest['happyScore'])

for k,row in richest.iterrows():
    plt.text(row['avg_income'],
             row['happyScore'],
             row['country'])
    
    # print(row['country'])
```

```python
from sklearn.cluster import KMeans
import numpy as np 

plt.xlabel('income')
plt.ylabel('happiness')

income_happiness = np.column_stack((income, happiness))

km_res = KMeans(n_clusters = 3).fit(income_happiness)

clusters = km_res.cluster_centers_

plt.scatter(income, happiness)
plt.scatter(clusters[:,0], clusters[:,-1], s = 1000, alpha = 0.75)
```
